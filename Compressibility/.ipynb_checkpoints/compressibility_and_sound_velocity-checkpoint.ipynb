## Calculate the velocity of pressure propagation using the fluid compressibility 

### Paraffin Density Equation 

```python
import numpy

def paraffin_density(pressure, temperature):
    return 8.33*(1.415e5+1.45e-4*pressure)/(1.802e5+212.3*(temperature-273.15))

def sound_velocity(density, compressibility):
    return numpy.sqrt(1.0/(density*compressibility))

def density_equation(pressure, compressibility):
    return density_0*(1.0/(1.0- (pressure-pressure_0)*compressibility))

```

```python
import scipy.optimize

pressures = numpy.logspace(4,8,100)
densities = paraffin_density(pressures, 330.0)

import matplotlib.pyplot as plt
%matplotlib inline

plt.plot(pressures, densities, "r")
plt.xlabel("Pressure [Pa]")
plt.ylabel("Density [$kg/m^3$]")
plt.title("Paraffin Density change with pressure")
plt.annotate(r"$ \rho[lb/gal]= 8.33 \left( \frac{1.415\times{10^{-5}}+1.45\times{10^{-4}} P[Pa]}{1.802\times{10^{-5}}+212.3(T[^oK]-273.15)} \right)  $" , xy=(0.15, 0.80), xycoords='figure fraction', size=16)

```

```python

density_0 = densities[0]
pressure_0 = pressures[0]

params, fitted = scipy.optimize.curve_fit(density_equation, pressures, densities, p0=[1e-8])
n_fit = params

qs = density_equation(pressures, n_fit)
plt.plot(pressures, densities, "ro", label="paraffin data")
plt.plot(pressures, qs, label="compressibility equation")
plt.xlabel("Pressure [Pa]")
plt.ylabel("Density [$kg/m^3$]")
plt.title("Fitting the compressibility equation with paraffin equation")
plt.annotate(r"$ \alpha= %5.4e [Pa^{-1}] $" % n_fit, xy=(0.5, 0.2), xycoords='figure fraction', size=18)
plt.legend(loc=0)
```

### Speed of sound

```python
sound_velocities = sound_velocity(densities*1000/8.34, n_fit)

plt.plot(pressures, sound_velocities, label="simulado")
plt.xlabel("Pressure [$Pa$]")
plt.ylabel("Velocity [$v/m$]")
plt.title("Speed of sound change with pressure")
plt.annotate(r"$ c = \sqrt{ \frac{1}{\alpha \rho} } $" , xy=(0.2, 0.3), xycoords='figure fraction', size=26)
plt.legend()
```

Speed of sound in Water

$ c_{water}[20^0C] = 1481.0 \frac{m}{s} $

Speed of sound in Solids (rocks)

$ c_{limestone} = 4500.0 \frac{m}{s} $

Speed of sound in Mixture

$ c_{mix} = c_{water}.y_{water}+c_{paraffin}.y_{paraffin}+c_{rock}.y_{rock} $
